```python
# Create model class
import torch
import torch.nn as nn

#  it is compulsory to inherit our module class with torch.nn's module to get access of all functionality and use it here.
class Model(nn.Module):
  def __init__(self, num_features):
    # then we will invoke constructor of parent class("nn.Module")
    super().__init__()
    # now we will create layer of our neural network, our this model will only have single layer, so will call it linear layer, this willbe our neuron
    self.linear = nn.Linear(num_features, 1)
    # activation function (sigmoid)
    self.sigmoid = nn.Sigmoid()

  def forward(self, features):
    out = self.linear(features)
    out = self.sigmoid(out)
    return out

```

```python
# create dataset
features = torch.rand(10,5)

# create model
model = Model(features.shape[1])

# Call model for forward pass
# model.forward(features)
# logically upper code should done, but pytorch recomendation is to do as bellow
# pytorch developer comunity have use magic class (__call__) that triggers forward() of class when someone called object of class.
model(features)
```

```python
# now we will create neural network with 1 hidden layer
#  it is compulsory to inherit our module class with torch.nn's module to get access of all functionality and use it here.
class Model(nn.Module):
  def __init__(self, num_features):
    # then we will invoke constructor of parent class("nn.Module")
    super().__init__()
    # now we will create layer of our neural network, our this model will only have single layer, so will call it linear layer, this willbe our neuron
    self.linear1 = nn.Linear(num_features, 3)
    # activation function (relu)
    self.relu = nn.ReLU()
    # second layer
    self.linear2 = nn.Linear(3, 1)
    self.sigmoid = nn.Sigmoid()

  def forward(self, features):
    out = self.linear1(features)
    out = self.relu(out)
    out = self.linear2(out)
    out = self.sigmoid(out)
    return out
```

```python
# create dataset
features = torch.rand(10,5)

# create model
model = Model(features.shape[1])

# Call model for forward pass
# model.forward(features)
# logically upper code should done, but pytorch recomendation is to do as bellow
# pytorch developer comunity have use magic class (__call__) that triggers forward() of class when someone called object of class.
model(features)
```

```python
# on the way to simplify this neural network we will use Sequential method of nn module
class Model(nn.Module):
  def __init__(self, num_features):
    super().__init__()
    self.network = nn.Sequential(
        nn.Linear(num_features, 3),
        nn.ReLU(),
        nn.Linear(3,1),
        nn.Sigmoid()
    );

  def forward(self, features):
    out = self.network(features)
    return out

# create dataset
features = torch.rand(10,5)

# create model
model = Model(features.shape[1])

# Call model for forward pass
# model.forward(features)
# logically upper code should done, but pytorch recomendation is to do as bellow
# pytorch developer comunity have use magic class (__call__) that triggers forward() of class when someone called object of class.
model(features)
```
